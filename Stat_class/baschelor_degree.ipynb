```python
# Re-importing necessary modules due to kernel reset
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the problem
p = 0.32  # population proportion
n = 500  # sample size
SE = np.sqrt(p * (1 - p) / n)  # standard error

# X values (Sample Proportions) range for graphing
x = np.linspace(0.25, 0.4, 100)
y = norm.pdf(x, p, SE)

# Function to plot and label specific regions with clear titles
def plot_distribution_with_title(x, y, x_shade, title, color):
    plt.figure(figsize=(8, 4))
    plt.plot(x, y, label='Sampling Distribution', color='blue')
    plt.fill_between(x_shade, norm.pdf(x_shade, p, SE), color=color, alpha=0.5)
    plt.title(title)
    plt.xlabel('Sample Proportion (p-hat)')
    plt.ylabel('Density')
    plt.legend(['Sampling Distribution'])
    plt.show()

# 1. Probability of 35% or More
x_35_or_more = np.linspace(0.35, 0.4, 100)
plot_distribution_with_title(
    x, y, x_35_or_more, 
    'Probability: Sample Proportion Greater Than 0.35', 'green'
)

# 2. Probability of 30% or Less
x_30_or_less = np.linspace(0.25, 0.30, 100)
plot_distribution_with_title(
    x, y, x_30_or_less, 
    'Probability: Sample Proportion Less Than 0.30', 'red'
)

# 3. Probability Between 28% and 38%
x_between_28_and_38 = np.linspace(0.28, 0.38, 100)
plot_distribution_with_title(
    x, y, x_between_28_and_38, 
    'Probability: Sample Proportion Between 0.28 and 0.38', 'purple'
)

# 4. Full Sampling Distribution
plot_distribution_with_title(
    x, y, x, 
    'Full Sampling Distribution of Sample Proportions (p-hat)', 'yellow'
)

```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Yes', 'No']
counts = [7, 43]

# Pie Chart
plt.figure(figsize=(8, 6))
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Individuals Divorced')
plt.show()
```

```python
import numpy as np
from scipy.stats import norm

# Plotting
z_values = np.linspace(-3, 3, 100)
plt.figure(figsize=(10, 6))
plt.plot(z_values, norm.pdf(z_values), label="Standard Normal Distribution")
plt.axvline(0.705, color='red', linestyle='--', label='z = 0.705')
plt.axvline(-0.705, color='red', linestyle='--')
plt.fill_between(z_values, 0, norm.pdf(z_values), where=(z_values >= 0.705), color='red', alpha=0.3)
plt.fill_between(z_values, 0, norm.pdf(z_values), where=(z_values <= -0.705), color='red', alpha=0.3)
plt.title('P-value for z-test')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Z-values for the curve
z_values = np.linspace(-3, 3, 100)
pdf = norm.pdf(z_values)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(z_values, pdf, label="Standard Normal Distribution")
plt.axvline(0.705, color='red', linestyle='--', label='z = 0.705')
plt.axvline(-0.705, color='red', linestyle='--')
plt.fill_between(z_values, 0, pdf, where=(z_values >= 0.705), color='red', alpha=0.3, label="p-value")
plt.fill_between(z_values, 0, pdf, where=(z_values <= -0.705), color='red', alpha=0.3)
plt.axhline(y=0, color='black', linewidth=0.5)
plt.title("Hypothesis Test: P-value vs Alpha")
plt.xlabel("z-values")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
p = 0.6  # population proportion
n = 500  # sample size
SE = np.sqrt(p * (1 - p) / n)  # standard error

# X values (Sample Proportions) range for graphing
x = np.linspace(0.5, 0.7, 100)

# Normal distribution for the sampling distribution
y = norm.pdf(x, p, SE)

# Function to plot and shade specific regions
def plot_distribution(x, y, x_shade, title, color):
    plt.figure(figsize=(8, 4))
    plt.plot(x, y, label='Sampling Distribution', color='blue')
    plt.fill_between(x_shade, norm.pdf(x_shade, p, SE), color=color, alpha=0.5)
    plt.title(title)
    plt.xlabel('Sample Proportion (p-hat)')
    plt.ylabel('Density')
    plt.legend(['Sampling Distribution'])
    plt.show()

# 1. Less than 58%
x_less_than = np.linspace(0.5, 0.58, 100)
plot_distribution(x, y, x_less_than, 'P(p-hat < 0.58)', 'red')

# 2. More than 63%
x_more_than = np.linspace(0.63, 0.7, 100)
plot_distribution(x, y, x_more_than, 'P(p-hat > 0.63)', 'green')

# 3. Between 58% and 63%
x_between = np.linspace(0.58, 0.63, 100)
plot_distribution(x, y, x_between, 'P(0.58 < p-hat < 0.63)', 'purple')

# 4. Full Distribution (Standard)
plot_distribution(x, y, x, 'Full Sampling Distribution', 'yellow')
```

```python
# Function to plot and label specific regions with clear titles
def plot_distribution_with_title(x, y, x_shade, title, color):
    plt.figure(figsize=(8, 4))
    plt.plot(x, y, label='Sampling Distribution', color='blue')
    plt.fill_between(x_shade, norm.pdf(x_shade, p, SE), color=color, alpha=0.5)
    plt.title(title)
    plt.xlabel('Sample Proportion (p-hat)')
    plt.ylabel('Density')
    plt.legend(['Sampling Distribution'])
    plt.show()

# 1. Less than 58%
x_less_than = np.linspace(0.5, 0.58, 100)
plot_distribution_with_title(
    x, y, x_less_than, 
    'Probability: Sample Proportion Less Than 0.58', 'red'
)

# 2. More than 63%
x_more_than = np.linspace(0.63, 0.7, 100)
plot_distribution_with_title(
    x, y, x_more_than, 
    'Probability: Sample Proportion Greater Than 0.63', 'green'
)

# 3. Between 58% and 63%
x_between = np.linspace(0.58, 0.63, 100)
plot_distribution_with_title(
    x, y, x_between, 
    'Probability: Sample Proportion Between 0.58 and 0.63', 'purple'
)

# 4. Full Sampling Distribution
plot_distribution_with_title(
    x, y, x, 
    'Full Sampling Distribution of Sample Proportions (p-hat)', 'yellow'
)
```
